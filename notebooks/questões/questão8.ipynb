```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vendas = pd.read_csv('../../datasets/vendas.csv')
produtos = pd.read_csv('../../datasets/produtos.csv')
```

```python
NAME = "GPS Garmin Vortex Maré Drift"
ID = 27
```

```python
matriz_interacao = pd.crosstab(vendas['client_id'], vendas['product_id'])
matriz_binaria = (matriz_interacao > 0).astype(int)
```

```python
similaridade_produtos = cosine_similarity(matriz_binaria.T)

df_sim = pd.DataFrame(
    similaridade_produtos, 
    index=matriz_binaria.columns, 
    columns=matriz_binaria.columns
)

ranking = df_sim[ID].drop(labels=[ID]).sort_values(ascending=False)
```

```python
top_5_ids = ranking.head(5).index
top_5_final = produtos[produtos['id'].isin(top_5_ids)].copy()

top_5_final['similaridade'] = top_5_final['id'].map(ranking)
```

```python
print(f"Produtos recomendados para quem comprou: {NAME}")

top_5_final[['id', 'name', 'similaridade']].sort_values(by='similaridade', ascending=False)
```
