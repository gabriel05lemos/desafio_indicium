## Importando Bibliotecas

```python
import pandas as pd
from matplotlib import pyplot as plt

pd.set_option('display.max_columns', None)
```

## Carregando Dados

```python
vendas = pd.read_csv('../../datasets/vendas.csv')
```

```python
vendas['sale_date'] = pd.to_datetime(vendas['sale_date'])

vendas.info()
```

## Criando Calendário

```python
datas = pd.date_range(start='2023-01-01', end='2024-12-31', freq='D')

calendario = pd.DataFrame({'data': datas})

mapa_dias = {
    'Monday': 'Segunda-feira',
    'Tuesday': 'Terça-feira',
    'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira',
    'Friday': 'Sexta-feira',
    'Saturday': 'Sábado',
    'Sunday': 'Domingo'
}

calendario['dia_da_semana'] = calendario['data'].dt.day_name().map(mapa_dias)

calendario.head()
```

## Fazendo Cálculos

```python
vendas_em_cada_dia = calendario.merge(
    vendas,
    left_on='data',
    right_on='sale_date',
    how='left'
)


vendas_por_dia = (
    vendas_em_cada_dia.
        groupby(['data', 'dia_da_semana'], as_index=False)
    ['total'].
        sum().
        sort_values('data')
)

vendas_por_dia.head()
```

```python
vendas_por_dia_da_semana_media = (
    vendas_por_dia.
        groupby('dia_da_semana')
    ['total'].
        mean().
        sort_values()
)

print(f"Dia da Semana com menor média de vendas: {vendas_por_dia_da_semana_media.index[0]}")

print(f"Média de vendas: {vendas_por_dia_da_semana_media.iloc[0]:,.2f}")
```

```python
vendas_por_dia_da_semana_media.plot(
    kind='bar',
    figsize=(12,6),
    title="Faturamento médio por Dia da Semana",
    rot=0,
    xlabel=''
)

plt.show()
```
