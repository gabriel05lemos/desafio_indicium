## Importando Bibliotecas

```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

## Carregando Dados

```python
produtos = pd.read_csv('../../datasets/produtos_raw.csv')

produtos.sample(10)
```

```python
produtos.info()
```

## Renomeando Colunas

As colunas `code` e `actual_category` serão renomeadas para garantir consistência e simplicidade.

```python
produtos.rename(
    columns={
        'code': 'id',
        'actual_category': 'category'
    },
    inplace=True
)

produtos.info()
```

## Analisando Duplicatas

```python
print(produtos.duplicated().sum())
```

```python
print(produtos.duplicated(subset='id').sum())
```

```python
print(produtos.duplicated(subset='name').sum())
```

```python
print(produtos.duplicated(subset='price').sum())
```

```python
produtos[produtos.duplicated(subset='price', keep=False)]
```

Temos 7 produtos duplicados, com as colunas `name`, `price` e `id` iguais. Vamos removê-los.

```python
produtos.drop_duplicates(
    subset='id',
    inplace=True
)

produtos.reset_index(
    drop=True,
    inplace=True
)

produtos.info()
```

## Padronizando Colunas

A coluna `price` não está com tipo numério e a coluna `category` possui formato fora de padrão. Vamos padronizar para poder normalizar os dados.

```python
def fix_category(row):
    product_id = row['id']
    if product_id <= 50:
        return 'eletrônicos'
    elif product_id  <= 100:
        return 'propulsão'
    else:
        return 'ancoragem'
```

```python
produtos['price'] = produtos['price'].apply(lambda x: x[3:]).astype('float')

produtos['category'] = produtos.apply(fix_category, axis=1)

produtos.info()
```

```python
produtos.sample(10)
```

## Normalização

Vamos criar uma `DataFrame` com as categorias e deixar apenas uma id numérica com os produtos.

```python
categorias = pd.DataFrame(
    {
    'id': [1, 2, 3],
    'name': ['eletrônicos', 'propulsão', 'ancoragem']
    }
)

produtos = (
    produtos.
        merge(
            categorias,
            left_on='category',
            right_on='name'
        ).drop(
            columns=['name_y', 'category']
        ).rename(
            columns={
                'name_x': 'name',
                'id_x': 'id',
                'id_y': 'category_id'
            }
        )
)
```

```python
produtos.sample(5)
```
