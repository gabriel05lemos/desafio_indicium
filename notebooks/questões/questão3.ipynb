## Importando Bibliotecas

```python
import pandas as pd
import json
```

## Carregando Arquivos

Vamos utilizar a biblioteca `json` para transformar o arquivo em um dicionário.

```python
with open("../../datasets/custos_importacao.json", 'r', encoding='utf-8') as f:
    dados = json.load(f)
```

## Transformando em DataFrame

Para transformar o dicionário aninhado em uma `Dataframe`, vamos utilizar a função nativa do Pandas `pd.json_normalize`.

```python
df = pd.json_normalize(
    dados, 
    record_path=['historic_data'], 
    meta=['product_id', 'product_name', 'category']
)

ordem_colunas = ['product_id', 'product_name', 'category', 'start_date', 'usd_price']
custos_importacao = df[ordem_colunas]
```

```python
custos_importacao.sample(10)
```

```python
custos_importacao.info()
```

## Corrigindo Tipos de Dados

As colunas `product_id` e `start_date` possuem tipos errados. Vamos corrigir os tipos.

```python
custos_importacao['product_id'] = custos_importacao['product_id'].astype('int')

custos_importacao['start_date'] =  pd.to_datetime(custos_importacao['start_date'], format='%d/%m/%Y')
```

```python
custos_importacao.info()
```

## Carregando o .CSV

```python
custos_importacao.to_csv('../../datasets/custos_importacao.csv', index=False, encoding='utf-8')
```
