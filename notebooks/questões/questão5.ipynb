```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
vendas = pd.read_csv('../../datasets/vendas.csv')

produtos = pd.read_csv('../../datasets/produtos.csv')

categorias = pd.read_csv('../../datasets/categorias.csv')
```

```python
vendas_com_categoria = vendas.merge(
    produtos[
        ['id', 'category_id']
        ],
    left_on='product_id',
    right_on='category_id'
)

dados_clientes = (
    vendas_com_categoria.
        groupby('client_id').
        agg({
            'total': 'mean',
            'category_id': 'nunique'
        })
)

dados_clientes.rename(
    columns={
        'total': 'ticket_medio',
        'category_id': 'diversidade'
    },
    inplace=True
)

dados_clientes
```

```python
clientes_fieis = (
    dados_clientes
        [dados_clientes['diversidade'] >= 3].
        sort_values('ticket_medio', ascending= False).
        head(10)
)

clientes_fieis
```

```python
ranking_categorias = (
    vendas_com_categoria
    [vendas_com_categoria['client_id'].isin(clientes_fieis.index)].
        groupby('category_id')
    ['qtd'].
        sum().
        sort_values(ascending=False)
)

ranking_categorias
```
