## Importandp Bibliotecas

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
```

## Carregando os Dados

```python
vendas = pd.read_csv('../../datasets/vendas.csv')
vendas['sale_date'] = pd.to_datetime(vendas['sale_date'])
```

## Selecionando o Produto

```python
motor_yamaha = vendas[vendas['product_id'] == 54].copy()
vendas_diarias = motor_yamaha.groupby('sale_date')['qtd'].sum().reset_index()
```

## Criando Calendário

```python
calendario = pd.date_range(start='2023-01-01', end='2024-01-31', freq='D')
modelo = pd.DataFrame({'sale_date': calendario})
modelo = pd.merge(modelo, vendas_diarias, on='sale_date', how='left').fillna(0)
```

## Criando Modelo

```python
modelo['previsao'] = modelo['qtd'].shift(1).rolling(window=7).mean()
```

```python
teste = modelo[(modelo['sale_date'] >= '2024-01-01') & 
                  (modelo['sale_date'] <= '2024-01-31')].copy()
```

```python
mae = mean_absolute_error(teste['qtd'], teste['previsao'])
```

```python
print(f"MAE para Janeiro de 2024: {mae:.0f} unidades")
```

## Calculando Previsões

```python
primeira_semana_janeiro = modelo[(modelo['sale_date'] >= '2024-01-01') & 
                  (modelo['sale_date'] <= '2024-01-07')]

print(f"Previsão de demanda para primeira semana de Janeiro de 2024: {primeira_semana_janeiro['previsao'].sum()} unidades")

print(f"Vendas da primeira semana de Janeiro de 2024: {primeira_semana_janeiro['qtd'].sum()} unidades")
```
